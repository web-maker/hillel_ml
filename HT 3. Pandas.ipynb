```python
import pandas as pd
```

```python
df = pd.read_csv("https://lms.ithillel.ua/api/lms/files/6337f0faf1ced7438c5f1899")
print(df.info())  # let's get some more information about the dataset
df.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
print(df["sex"].value_counts(), "\n")
print("There are 21790 men and 10771 women in this dataset")
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
mean_age = df[df["sex"] == "Female"]["age"].mean()
print(f"Average age of women is {int(mean_age.round(0))} years")
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
# First of all, let's see which values are there
print(df["native-country"].value_counts(normalize=False), "\n")

# We can see that there are 137 German citizens, but let's show this in a more readable format
counted = df["native-country"].value_counts(normalize=True).mul(100).round(1).astype(str) + "%"
print("The total percentage of German citizens is", counted["Germany"])
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
# First of all, let's check which values we have for salary in the dataset
print(df["salary"].value_counts(), "\n")

# Okay, now we can get all people, that have a salary >50K
more50k = df[df["salary"] == ">50K"]

print("Age mean with the salary >50K:", more50k["age"].mean().round(0).astype(int), "years")
print("Standard deviations of the age for peoples with salary >50K:", more50k["age"].std(), "\n")

# And let's have the same metrics for people with salary <=50K
less50k = df[df["salary"] == "<=50K"]

print("Age mean with the salary <=50K:", less50k["age"].mean().round(0).astype(int), "years")
print("Standard deviations of the age for peoples with salary <=50K:", less50k["age"].std())
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
# Let's count educations gradations for people with a salary >50K
more50k["education"].value_counts()
print(more50k.groupby("education")["education"].count(), "\n")

# So, as we can see, it's not a true, because there is lot of peoples with Some-college and HS-grad education
# But let's calculate the percentage of high-level education
with_highlevel_ed = more50k[more50k["education"].isin(["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"])].shape[0]
print("Total number of poples with salary >50K:", more50k.shape[0])
print("Total number of peoples with_highlevel education:", with_highlevel_ed)
print("Percentage value of peoples with highlevel education:", round(with_highlevel_ed * 100 / more50k["education"].shape[0], 2), "%")
# Answer - no, it's not true. For the provided dataset, only 57.84% of people with a salary >50K had a high-level education
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
# First of all, let's print a general statistic
df.groupby(["race", "sex"]).describe()
```

```python
# As we can see, max age for men of the Amer-Indian-Eskimo race is 82. Let's print it.
print("Max age for Amer-Indian-Eskimo mens is:", df[(df["race"] == "Amer-Indian-Eskimo") & (df["sex"] == "Male")]["age"].max())
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Let's add new a feature `is_married`
# We can create a new column with `.apply`, like below:
# more50k["is_married"] = df["marital-status"].apply(lambda x: True if x.startswith("Married") else False)
# Or we can use a mask here:
# more50k["is_married"] = pd.Series(False, index=more50k.index).mask(df["marital-status"].str.startswith("Married"), True)
# But both these solutions produce a SettingWithCopyWarning.
# So, let's use `.loc` (this one does not produce any warnings)

more50k = more50k.assign(is_married=False)
more50k.loc[more50k["marital-status"].str.startswith("Married"), "is_married"] = True
more50k[(more50k["sex"] == "Male")]["is_married"].value_counts(normalize=True).mul(100).round(1).astype(str) + "%"

# Answer - of all men with a salary >50k, married men have a bigger proportion (89.5%)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Calculate a number of max working hours first
max_hours = df["hours-per-week"].max()

# Now we can filter all the crazy peoples which are work the same number of hours per week
crazy_peoples = df[df["hours-per-week"] == max_hours]

# And we can show a result
print(f"Number of peoples that are working {max_hours} hours in week:", len(crazy_peoples))
print("From them number of peoples which get more >50k salary:", len(crazy_peoples[crazy_peoples["salary"] == ">50K"]))
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
print("Mean hours per week for >50k salary:")
more50k.groupby("native-country")["hours-per-week"].mean()
```

```python
print("Mean hours per week for <=50K salary:")
less50k.groupby("native-country")["hours-per-week"].mean()
```

```python
# Task 1 - improve dataframe types
print("Original DataFrame info:", df.info(), "\n")
initial_memory_usage = df.memory_usage(deep=False).sum()  # store non-deep memory usage for dataframe in bytes

# So, memory usage now is 3.7+ MB
# Let's make some improvements
print("Max age value:", df["age"].max())
df["age"] = df["age"].astype("int8")  # max age is 90, so we can use int8 here, because int8 can represents 2^8 (256) value

# Now let's check fnlwgt value
print("Max fnlwgt value:", df["fnlwgt"].max())
df["fnlwgt"] = df["fnlwgt"].astype("int32")  # we can use int32 here instead of int64

# Let's improve education-num
print("Max education-num value:", df["education-num"].max())  # max education-num is 16, so le's use int8 here
df["education-num"] = df["education-num"].astype("int8")

# capital-gain now
print("Max capital-gain value:", df["capital-gain"].max())
df["capital-gain"] = df["capital-gain"].astype("int32")  # we can use int32 here, because max value for capital-gain is 99999

# capital-loss
print("Max capital-loss value:", df["capital-loss"].max())
df["capital-loss"] = df["capital-loss"].astype("int16")  # int16 can be used here, because max value here is 4356

# Let's check hours-per-week
print("Max hours-per-week value:", df["hours-per-week"].max(), "\n")
df["hours-per-week"] = df["hours-per-week"].astype("int8")  # max value here is 99, so let's use int8

# Show results
print("Optimized memory usage:", (initial_memory_usage - df.memory_usage(deep=False).sum()) / 1024 / 1024, "MB")

# Conclusion - data type improvements can significantly reduce memory usage.
# Even minor updates, like changing data types for int columns can free a useful part of memory, especially for the large datasets
```

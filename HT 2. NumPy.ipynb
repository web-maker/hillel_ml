```python
# Import numpy
import numpy as np
```

```python
# Load dataset
# uncomment code below if you want to load dataset from mounted Goodle Drive directory
# data = np.loadtxt("/content/drive/MyDrive/Hillel ML 09.06.2023/datasets/archive.ics.uci.edu_ml_machine-learning-databases_iris_iris.data.txt", delimiter=",", dtype="str")
# data  # uncomment this variable if you want to check entire dataset
data = np.genfromtxt("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", dtype=str, delimiter=",")
```

```python
# Let's check dataset parameters
print("Shape: ", data.shape)
print("Datatype: ", data.dtype)
print("Size: ", data.size)
print("Dimensions: ", data.ndim)
```

```python
# Task 1 - delete a column with a target
# delete method returns a new array, so let's store it into a new variable `cleared_data`
cleared_data = np.delete(data, 4, axis=1)
# cleared_data  # uncomment this variable if you want to check entire dataset
# check the shape and size of the cleaned dataset
print("Shape: ", cleared_data.shape)
print("Size: ", cleared_data.size)
```

```python
# Task 2 - Transform columns that are left in a 2D array (or check, if it’s still a 2D array)
print("Dimensions: ", cleared_data.ndim)
# Yep, we are still in 2D array, because we were using np.delete for removing last column. Otherwise, dataset should be reshaped.
```

```python
# Task 3 - Calculate mean, median and standard deviation for the 1-st column
# First of all let's convert dataset to a float data
cleared_data = cleared_data.astype("float32")
first_column_data = cleared_data[:, 0]
print("Mean: ", np.mean(first_column_data))
print("Median: ", np.median(first_column_data))
print("Standard deviation: ", np.std(first_column_data))
```

```python
# Task 4 - Insert 20 np.nan on a random positions in the array
np.random.seed(100)
indices = np.random.choice(range(cleared_data.size), size=20, replace=False)  # get 20 unique positions
print(indices)

original_shape = cleared_data.shape  # store original array shape here
cleared_data.ravel()  # we don't need a copy here, so let's use reveal() instaed of flatten()

np.put(cleared_data, indices, np.nan)  # replace values by indices
print(np.take(cleared_data, indices))  # let's re-check if all indices were replaced correctly

# and, finally, let's reshape data to the original shape
np.reshape(cleared_data, original_shape)
print("Dimensions: ", cleared_data.ndim)
```

```python
# Task 5 - Find positions of inserted NaN values in the 1st column
indeces = np.argwhere(np.isnan(cleared_data[:, 0]))
print(indeces)

# recheck if we found the indices correctly
print(cleared_data[:, 0][12])
print(cleared_data[:, 0][90])
print(cleared_data[:, 0][106])
```

```python
# Task 6 - Filter array by conditions: values in 3rd column > 1.5 and values in the 1st column < 5.0 (save into a new variable)
mask = np.logical_and(cleared_data[:, 2] > 1.5, cleared_data[:, 0] < 5.0)
filtered_data = cleared_data[mask]
filtered_data
```

```python
# Task 7 - Replace all np.nan with 0
# There is, at least, 2 ways to do that
# Option 1 - use basic syntax + np.isnan
a = np.copy(cleared_data)
a[np.isnan(a)] = 0
# print(a)
# Option 2 - use np.nun_to_num(0)
np.nan_to_num(cleared_data, copy=False, nan=0)
print(cleared_data)
```

```python
# Task 8 - Count all unique values in the array and output them together with their quantity
values, counts = np.unique(cleared_data, return_counts=True)
print("Values : ", values)
print("Counts : ", counts)
```

```python
# Task 9 - Split an array into multiple sub-arrays horizontally into 2 parts
# Interesting, but looks like hsplit makes vertical split, but not horizantal.
# And vsplit makes a horizontal split.
part_1, part_2 = np.vsplit(cleared_data, 2)
print(part_1.shape)
print(part_2.shape)
# or we can use np.split. Basically, vsplit it's equivalent to split with axis=0
part_1, part_2 = np.split(cleared_data, 2, axis=0)
print(part_1.shape)
print(part_2.shape)
```

```python
# Task 10 - Sort both arrays by 1-st column: 1-st array ascending order, 2-nd by descending
# sort first array in ascending order by the first column
part_1_sorted_indeces = part_1[:, 0].argsort()
part_1 = part_1[part_1_sorted_indeces]
print(part_1_sorted_indeces)
# print(part_1)

# sort seconds array in descending order by the first column
part_2_sorted_indeces = part_2[:, 0].argsort()
part_2 = part_2[part_2_sorted_indeces[::-1] ]  # desc sorting
print(part_2_sorted_indeces)
# print(part_2)
```

```python
# Task 11 - Concatenate array into one
# np.concatenate([part_1, part_2], axis=0) or
# np.r_[part_1, part_2] also can be used
final_data = np.vstack([part_1, part_2])
print(final_data)
```

```python
# Task 12 - Find most frequent number in array
values, counts = np.unique(final_data, return_counts=True)
print("Most frequent values is: ", values[counts.argmax()])
```

```python
# Task 13 - Create a function, which should multiply on 2 all values in the column,
# which are less than mean value of this column, and should divide all other values on 4
def modify(arr):
  """Modify provided array."""
  return np.where(arr < np.mean(arr), arr * 2, arr / 4)
```

```python
# Task 14 - Apply function to the 3-rd column
final_data[:, 2] = modify(final_data[:, 2])
print(final_data)
```
